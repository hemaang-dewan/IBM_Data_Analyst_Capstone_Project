## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Change the figure size of the seaborn plot
sns.set(rc={"figure.figsize":(18, 10)})
sns.scatterplot(data=df, x='Age', y='WorkWeekHrs')

plt.title('WorkWeekHrs by Age', size=16)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
sns.set(rc={"figure.figsize":(18, 10)})
sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5, sizes=(10, 500))

plt.title('WorkWeekHrs and CodeRevHrs by Age', size=16)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(Respondent) AS 'Number of Respondents'
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(Respondent) DESC
LIMIT 5
"""

top5_dbs_df = pd.read_sql_query(QUERY,conn)
top5_dbs_df = top5_dbs_df.set_index('DatabaseDesireNextYear')

colors_list = ['#345c33', '#55752b', '#828a19', '#bc9c00', '#ffa600']

plt.figure(figsize=(18, 10))
plt.pie(x='Number of Respondents', data=top5_dbs_df, labels=top5_dbs_df.index, colors=colors_list, autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Want to Learn Next Year')
plt.legend(labels=top5_dbs_df.index, loc='upper right')
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
"""
df = pd.read_sql_query(QUERY,conn)

work_week_hrs_median = df['WorkWeekHrs'].quantile(0.5)
code_rev_hrs_median = df['CodeRevHrs'].quantile(0.5)
age = df['Age']

plt.figure(figsize=(18, 10))

fig, ax = plt.subplots()
ax.bar(age, work_week_hrs_median, label='WorkWeekHrs')
ax.bar(age, code_rev_hrs_median, bottom=work_week_hrs_median, label='CodeRevHrs')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Respondents Between the Ages of 30 and 35')
ax.set_xlabel('Age')
ax.set_ylabel('Number of Hours')
ax.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df = pd.read_sql_query(QUERY,conn)
df = df.groupby('Age').median()

plt.figure(figsize=(18, 10))
plt.plot(df)
plt.title('Median ConvertedComp of Respondents from Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) as Frequency
FROM master
GROUP BY MainBranch
"""

df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(18, 10))
plt.barh(y='MainBranch', width='Frequency', data=df)
plt.title('MainBranch Categories by Frequency')
plt.xlabel('Frequency')
plt.ylabel('MainBranch')

plt.show()
```
